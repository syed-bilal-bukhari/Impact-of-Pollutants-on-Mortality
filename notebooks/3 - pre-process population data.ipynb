## 1. Import Required Libraries

We'll start by importing pandas, which is essential for data manipulation and analysis.

## 2. Load Datasets


Loading CSV file for the Population Data. The data is sourced from this World Bank dataset which in turn lists as sources: 
- United Nations Population Division. World Population Prospects
- United Nations Statistical Division. Population and Vital Statistics Reprot (various years)
- Census reports and other statistical publications from national statistical offices

```python
# Read the CSV file
file_path = "../Datasets/world_bank_population_data.csv"
population_data = pd.read_csv(file_path)
```

## 3. Initial Data Exploration


Let's preview the first few rows from dataset to understand its structure and content. This will help us verify that the data was loaded correctly.

```python
population_data.head()
```

### 3.2 Analyzing Column Contents


This analysis examines key columns
1. The number of unique countries
2. The time range of dataset

This helps identify consistent patterns, variations, and potential issues across datasets before merging them.

```python
unique_countries = population_data['Country Name'].nunique()
print(f"Number of unique countries: {unique_countries}")
```

```python
min_year = population_data['Year'].min()
max_year = population_data['Year'].max()

print(f"Minimum Year: {min_year}, Maximum Year: {max_year}")
```

Since Our analysis is based from 1990 to 2020, We would be dropping rest of the rows.

```python
population_data = population_data[(population_data['Year'] >= 1990) & (population_data['Year'] <= 2020)]
population_data.reset_index(drop=True, inplace=True)
```

```python
min_year = population_data['Year'].min()
max_year = population_data['Year'].max()

print(f"Minimum Year: {min_year}, Maximum Year: {max_year}")
```

We know that in our Pollutants Data and Cause of Death data, the country's name is represented by the column Country, so we will rename it right here before exporting

```python
population_data.rename(columns={'Country Name': 'Country', 'Value': 'Population'}, inplace=True)
```

## 5. Feature Selection

Following the data analysis, we're removing columns identified as redundant or unnecessary for our analytical goals. This includes:
- Country Code: Removed, as the country name already provides sufficient information.

This streamlines the dataset for more focused analysis.

```python
population_data.drop(columns=['Country Code'], inplace=True)
population_data.head()
```

## 6. Missing Value Detection and Handling:
Upon checking the Missing Values, there is no row that contain any null or missing value

```python
# Show count of null/NA values for each column in merged_data
null_counts = population_data.isnull().sum()
print("Null/NA values per column:")
print(null_counts)
```

```python
output_file_path = "./processed_population_data.csv"
population_data.to_csv(output_file_path, index=False)
print(f"Processed population data exported to {output_file_path}")
```
